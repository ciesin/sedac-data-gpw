```python
feature_fields_list = arcpy.ListFields(dashboard_shape_path)
```

```python
field_alias_list=[]
for field in feature_fields_list:
    field_alias_list.append(field.name)
field_alias_list= field_alias_list[6:]
field_alias_list = ['ISO',
 'ISO_alpha3_Code',
 'Global_Code',
 'Global_Name',
 'Region_Code',
 'Region_Name',
 'Sub_region_Code',
 'Sub_region_Name',
 'Intermediate_Region_Code',
 'Intermediate_Region_Name',
 'Country_or_Area',
 'M49_Code__AKA_ISO_Numeric_',
 'ISO_alpha2_Code',
 'Continent_GPW',
 'Least_Developed_Countries__LDC_',
 'Land_Locked_Developing_Countries__LLDC_',
 'Small_Island_Developing_States__SIDS_',
 'CensusTableRowsPerCountry',
 'Stage_s_',
 'Status',
 'LatestCensusYear',
 'PreviousCensusYear',
 'Collection_Type',
 'Frequency',
 'Variable_Ingested',
 'Hightest_Admin_Level_Ingested',
 'Download_Link']
field_alias_list
```

```python
with arcpy.da.UpdateCursor(dashboard_shape_path, field_alias_list) as cursor:
    # For each row, evaluate the WELL_YIELD value (index position 
    # of 0), and update WELL_CLASS (index position of 1)
    for row in cursor:
        iso_val = row[0]#'ISO',
        if iso_val in update_dict:
            print(iso_val)
            #print(update_dict[iso_val]['Stage(s)'])
            row[1]  = update_dict[iso_val]['ISO-alpha3 Code']# 'ISO_alpha3_Code',
            row[2] = update_dict[iso_val]['Global Code']# 'Global_Code',
            row[3]= update_dict[iso_val]['Global Name']# 'Global_Name',
            row[4]= update_dict[iso_val]['Region Code']# 'Region_Code',
            row[5]= update_dict[iso_val]['Region Name']# 'Region_Name',
            row[6]= update_dict[iso_val]['Sub-region Code']# 'Sub_region_Code',
            row[7]= update_dict[iso_val]['Sub-region Name']# 'Sub_region_Name',
            row[8]= update_dict[iso_val]['Intermediate Region Code']# 'Intermediate_Region_Code',
            row[9]= update_dict[iso_val]['Intermediate Region Name']# 'Intermediate_Region_Name',
            row[10]= update_dict[iso_val]['Country or Area']# 'Country_or_Area',
            row[11]= update_dict[iso_val]['M49 Code (AKA ISO Numeric)']# 'M49_Code__AKA_ISO_Numeric_',
            row[12]= update_dict[iso_val]['ISO-alpha2 Code']# 'ISO_alpha2_Code',
            row[13]= update_dict[iso_val]['Continent GPW']# 'Continent_GPW',
            row[14]= update_dict[iso_val]['Least Developed Countries (LDC)']# 'Least_Developed_Countries__LDC_',
            row[15]= update_dict[iso_val]['Land Locked Developing Countries (LLDC)']# 'Land_Locked_Developing_Countries__LLDC_',
            row[16]= update_dict[iso_val]['Small Island Developing States (SIDS)']# 'Small_Island_Developing_States__SIDS_',
            row[17]= update_dict[iso_val]['CensusTableRowsPerCountry']# 'CensusTableRowsPerCountry',
            row[18]= update_dict[iso_val]['Stage(s)']# 'Stage_s_',
            row[19]= update_dict[iso_val]['Status']# 'Status',
            row[20]= update_dict[iso_val]['LatestCensusYear']# 'LatestCensusYear',
            row[21]= update_dict[iso_val]['PreviousCensusYear']# 'PreviousCensusYear',
            row[22]= update_dict[iso_val]['Collection Type']# 'Collection_Type',
            row[23]= update_dict[iso_val]['Frequency']# 'Frequency',
            row[24]= update_dict[iso_val]['Variable Ingested ']# 'Variable_Ingested',
            row[25]= update_dict[iso_val]['Hightest Admin Level Ingested ']# 'Hightest_Admin_Level_Ingested',
            row[26]= update_dict[iso_val]['Download Link']# 'Download_Link'
            cursor.updateRow(row)
 
 
        
```
